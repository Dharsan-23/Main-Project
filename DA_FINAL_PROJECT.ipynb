## Handle duplicates

```python
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_rows}")

if  duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print(f"Duplicate rows removed. ")
else:
    print("No duplicate.")
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'])

df['Join_Date'] = df['Join_Date'].dt.strftime('%d/%m/%Y')

print(df['Join_Date'].head())
```

```python
df
```

```python
df['Phone'] = df['Phone'].abs()
```

```python
df
```

# **Treat outliers if required**

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3-Q1

Lower_Value = Q1 - 1.5 * IQR
Upper_Value = Q3 + 1.5 * IQR

Outliers = df[(df['Age']< Lower_Value) | (df['Age']> Upper_Value)]

print(Outliers)
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3-Q1

Lower_Value = Q1 - 1.5 * IQR
Upper_Value = Q3 + 1.5 * IQR

Outliers = df[(df['Salary']< Lower_Value) | (df['Salary']> Upper_Value)]

print(Outliers)
```

# **Check skewness and apply transformations**

```python
df['Salary_Log'] = np.log1p(df['Salary'])
print("New Salary Skewness : ", df['Salary_Log'].skew())
```

```python
df['Age_Log'] = np.log1p(df['Age'])
print("New Age Skewness : ", df['Age_Log'].skew())
```

```python
df['Salary_Sqrt'] = np.sqrt(df['Salary'])
print('Sqrt of Salary : ', df['Salary_Sqrt'])
```

```python
df['Age_Sqrt'] = np.sqrt(df['Age'])
print('Sqrt of Age : ', df['Age_Sqrt'])
```
